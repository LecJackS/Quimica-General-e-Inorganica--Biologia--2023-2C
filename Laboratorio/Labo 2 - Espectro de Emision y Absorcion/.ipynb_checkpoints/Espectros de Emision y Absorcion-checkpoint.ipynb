```python
import pandas as pd
import plotly.express as px
```

```python
emi_data_path = "./emi.csv"

emi_data = pd.read_csv(emi_data_path, header=1, delimiter=";")
```

```python
emi_data.head()
```

```python
# Grafico de puntos conectados
integracion = 1000 # en ms
fig = px.line(emi_data, x=f'lambda_{integracion}', y=f'intensity_{integracion}',
              title=f'Lambda vs. Intensidad - Integración {integracion}ms',
              labels={f'lambda_{integracion}': 'Longitud de onda (nm)',
                      f'intensity_{integracion}': 'Intensidad (recuento)'})

# Agrego mediciones como círculos
fig.add_scatter(x=emi_data[f'lambda_{integracion}'],
                y=emi_data[f'intensity_{integracion}'],
                mode='markers', name='Medición')

# Agrego líneas verticales en los valores de las líneas espectrales más frecuentes
vertical_lines = [410, 434, 486, 657]
for value in vertical_lines:
    fig.add_vline(x=value, line_dash='dash', line_color='green', line_width=1, name=f'Line at {value}')

# Cambio los valores numericos que se muestran en el eje x
x_axis_tickvals = vertical_lines + list(range(350, 1000, 200))
x_axis_tickvals.sort()  # Sort the tick values for clarity
fig.update_xaxes(tickvals=x_axis_tickvals)

# Cambio el color del fondo a blanco y ejes
fig.update_layout(plot_bgcolor='white')
# add x axis tick lines
fig.update_xaxes(tickvals=x_axis_tickvals, showline=True, linewidth=1, linecolor='black', ticks="outside")
fig.update_yaxes(showline=True, linewidth=1, linecolor='black', ticks="outside", gridcolor='lightgray')

fig.show()
```

```python
import numpy as np
```

```python

lambdas = np.asarray([657, 486, 434, 410])

print("lambdas (nm):", lambdas)
print("lambdas (m):", lambdas * 1e-9)

lambdas_inv = 1 / (lambdas * 1e-9)
print("lambdas_inv (1/m):", lambdas_inv)
```

```python
ns = np.asarray([3, 4, 5, 6])
ns2_inv = 1 / (ns**2)
ns2_inv
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

model = LinearRegression().fit(ns2_inv.reshape(-1, 1), lambdas_inv)
y_hat = model.predict(ns2_inv.reshape(-1, 1))
y_hat
```

```python
model.coef_[0]*1e-7
```

```python
r2_score(lambdas_inv, y_hat)
```

```python
fig = px.line(x=ns2_inv, y=lambdas_inv,
              title=r"$\large{\frac 1 {\lambda} \ \text{vs.}\  \frac 1 {n^2}} \text{ - Verificación de Constante de Rydberg - } R^2 = 0.99999889992$",
              labels={'x': r'$\large \frac 1 {n^2}$',
                      'y': r'$\large \frac 1 {\lambda} \left(\frac 1 m \right)$'})


# Agrego predicciones lineales como círculos grandes
fig.add_scatter(x=ns2_inv,
                y=y_hat,
                mode='markers', name='Predicción de ajuste lineal',
                marker=dict(
                    color='white',
                    size=12,
                    line=dict(
                        color='MediumPurple',
                        width=1
                    )
                )
               )
# Agrego mediciones como círculos
fig.add_scatter(x=ns2_inv,
                y=lambdas_inv,
                mode='markers', name='Medición')

x_axis_tickvals = []
x_axis_tickvals += list(ns2_inv.round(3))
# Cambio el color del fondo a blanco y ejes

# add x axis tick lines
fig.update_xaxes(tickvals=x_axis_tickvals, showline=True, linewidth=1, linecolor='black', ticks="outside")
fig.update_yaxes(showline=True, linewidth=1, linecolor='black', ticks="outside", gridcolor='lightgray')
fig.update_layout(plot_bgcolor='white')

fig.show()
```
